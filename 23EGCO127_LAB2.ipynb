```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Recursively traversing dataset directories to load CSV files and dynamically injecting their folder
main_dir = "datasets/SmartGrid-CyberPhysical-Attack-Dataset/Dataset/"


data_frames = []
for dirpath, dirnames, filenames in os.walk(main_dir):
    root_dir = Path(dirpath)

    # Extract the folder name to use as the label (e.g., 'benign', 'Bruteforce')
    # root_dir.name gives just the last folder in the path
    label = root_dir.name
    print(label)

    for file in filenames:
        # Ensure we are only reading CSV files to avoid errors
        if file.endswith('.csv'):
            file_path = root_dir / file

            # Added low_memory=False to resolve DtypeWarning
            df = pd.read_csv(file_path, low_memory=False)

            # Assign the label to a new column named 'label'
            df['label'] = label
            data_frames.append(df)
```

```python
if data_frames:
    df = pd.concat(data_frames, ignore_index=True)
    print("Successfuly combined all CSV files.")
else:
    print("No CSV files found.")
```

```python
# Drop totally empty columns
columns_to_drop = ['Unnamed: 12', 'udp.srcport', 'udp.dstport']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
# fixing inconsistent data. types
string_numeric_cols = ['V_A', 'V_B', 'V_C', 'I_A', 'I_B', 'I_C',
'ActivePower', 'ReactivePower', 'ip.len']
for col in string_numeric_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace(',', '')
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Verify Dtypes
df.info()
```

```python
# Handling duplicate entries
duplicates = df.duplicated().sum()
print(f"Removing {duplicates} duplicate rows...")
df.drop_duplicates(inplace=True)
```

```python
# Safely handling missing values
categorical_cyber = ['eth.src', 'eth.dst', 'ip.src', 'ip.dst',
'_ws.col.Protocol', 'frame.time_epoch']
for col in categorical_cyber:
    if col in df.columns:
        df[col] = df[col].fillna('UNKNOWN')
```

```python
# imputing continuous numerical features
numeric_cols = df.select_dtypes(include=[np.number]).columns
cols_to_impute = [col for col in numeric_cols if col not in ['node',
'PLCID']]
for col in cols_to_impute:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
# Drop rows only if both 'node' and 'PLCID' identifiers are missing to preserve complementary cyber-physical data

df.dropna(subset=['node', 'PLCID'], how='all', inplace=True)
print(f"Dataset shape after safe drop: {df.shape}")
```

```python
# Dropping network packet metadata, timestamps, and target labels to prevent model overfitting
cols_to_exclude = ['frame.time_epoch', '@timestamp', 'label',
'eth.src', 'eth.dst', 'ip.src', 'ip.dst', '_ws.col.Protocol']

X_features = df.drop(columns=[col for col in cols_to_exclude if col in
df.columns])
y_target = df['label']
print(f"Selected Predictor Features ({X_features.shape[1]})")
print(f"Columns: \n{list(X_features.columns)}")

```

```python
# labelEncoder for algorithmic compatibility

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_target)
class_mapping = dict(zip(range(len(label_encoder.classes_)),
label_encoder.classes_))
print(f"\nTarget Classes Encoded: {class_mapping}")
```

```python
# Partition the dataset into an 80% training set and a 20% testing set using stratified sampling to preserve original class distribution
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_encoded, test_size=0.2, random_state=42,
stratify=y_encoded
)
```

```python
# Generate descriptive summary statistics for the training features to analyze their baseline data distributions before applying scaling transformation
X_train.describe().T
```

```python
#Standardise the training and testing feature matrices using a StandardScaler fitted exclusively on the training partition to eliminate feature magnitude bias and prevent data leakage.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Convert the scaled training array back into a DataFrame to verify that all features now have a mean of approximately 0 and a standard deviation of 1.
X_train_scaled_df = pd.DataFrame(X_train_scaled,
columns=X_train.columns)
X_train_scaled_df.describe().T
```
